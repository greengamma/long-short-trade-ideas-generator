```python
# Addititive Decomposition (y = Trend * Seasonal * Residuals)
result_add = seasonal_decompose(df[ratio], model="additive",period = 30)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)
result_add.plot();
```

```python
# Multiplicative Decomposition (y = Trend * Seasonal * Residuals)
result_mul = seasonal_decompose(df[ratio], model="multiplicative",period = 30)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,8)
result_mul.plot();
```

```python
# Plot the residuals with "result_add.resid" to decide
f, (ax1, ax2) = plt.subplots(1,2, figsize=(13,3))
ax1.plot(result_add.resid); ax1.set_title("Additive Model Residuals")
ax2.plot(result_mul.resid); ax2.set_title("Multiplicative Model Residuals");
```

```python
# Build model
arima = ARIMA(df_train, order=(0, 1, 0), trend='t')  
arima = arima.fit()

## Forecast
# Forecast values
forecast = arima.forecast(len(df_test), alpha=0.05)  # 95% confidence

# Forecast values and confidence intervals
forecast_results = arima.get_forecast(len(df_test), alpha=0.05)
forecast = forecast_results.predicted_mean
confidence_int = forecast_results.conf_int()
```

```python
forecast_df = pd.DataFrame(forecast)
forecast_df["id"] = list(df_test.index)
forecast_df.set_index("id",inplace=True)
```

```python
confidence_int["id"] = list(df_test.index)
confidence_int.set_index("id",inplace=True)
confidence_int= np.array(confidence_int)
```

```python
# We define here a "Plot forecast vs. real", which also shows historical training set

def plot_forecast(fc, train, test, upper=None, lower=None):
    is_confidence_int = isinstance(upper, np.ndarray) and isinstance(lower, np.ndarray)
    # Prepare plot series
    #fc_series = pd.Series(fc, index=test.index)
    
    lower_series = pd.Series(upper,index=test.index) if is_confidence_int else None
    upper_series = pd.Series(lower,index=test.index) if is_confidence_int else None
    
    # Plot
    plt.figure(figsize=(10,4), dpi=100)
    plt.plot(train, label='training', color='black')
    plt.plot(test, label='actual', color='black', ls='--')
    plt.plot(fc, label='forecast', color='orange')
    if is_confidence_int:
        plt.fill_between(lower_series.index, lower_series, upper_series, color='k', alpha=.15)
    plt.title('Forecast vs Actuals')
    plt.legend(loc='upper left', fontsize=8);
```

```python
plot_forecast(forecast_df, df_train, df_test, confidence_int[:,0], confidence_int[:,1])
```

```python
# Re-compose back to initial TS

forecast_recons = forecast_df["predicted_mean"] * result_mul.seasonal[(len(df_train)):]
train_recons = df_train[ratio] * result_mul.seasonal[0:(len(df_train))]
test_recons = df_test[ratio] * result_mul.seasonal[(len(df_train)):]
lower_recons = confidence_int[:, 0] * result_mul.seasonal[(len(df_train)):]
upper_recons = confidence_int[:, 1] * result_mul.seasonal[(len(df_train)):]

# Plot
plot_forecast(forecast_recons, train_recons, test_recons, lower_recons.values, upper_recons.values)
```

```python
mape_arima_deseasonalized = np.mean(np.abs(forecast_recons - test_recons)/np.abs(test_recons))
mape_arima_deseasonalized
```

# ARIMA Summary


```python
ratio_list = list(df_init.columns)
ratio_list
```

```python
results = []

for i in range(len(ratio_list)):
    
    #train and test split
    
    ratio = [ratio_list[i]]
    
    df = df_init[ratio_list[i]]
    train_size = 0.6
    index = round(train_size*df.shape[0])

    df_train = df.iloc[:index]
    df_test = df.iloc[index:]
    
    #auto - ARIMA
    model = pm.auto_arima(df,
                       start_p=1, max_p=6,
                       start_q=1, max_q=6,
                       trend='t',
                       max_d = 2,
                       seasonal=False,
                       trace=True)
    
    # Build model
    arima = ARIMA(df_train, order=model.get_params().get("order"), trend='t')  
    arima = arima.fit()

    ## Forecast
    # Forecast values
    forecast = arima.forecast(len(df_test), alpha=0.05)  # 95% confidence
    
    forecast_df = pd.DataFrame(forecast)
    forecast_df["id"] = list(df_test.index)
    forecast_df.set_index("id",inplace=True)
    
    #append to MAPE list
    mape_arima = np.mean(np.abs(forecast_df["predicted_mean"] - df_test)/np.abs(df_test))
    results.append(mape_arima)
    
summary = pd.DataFrame()
summary["ratio"] =  ratio_list
summary["MAPE"] = results
summary
```
